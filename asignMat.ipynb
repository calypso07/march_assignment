Bar plot - A bar plot is used to compare values across different categories. It is created using the bar() function in Matplotlib.

Scatter plot - A scatter plot is used to show the relationship between two variables. It is created using the scatter() function in Matplotlib.

Histogram - A histogram is used to show the distribution of a single variable. It is created using the hist() function in Matplotlib.

Box plot - A box plot is used to show the distribution of a variable. It is created using the boxplot() function in Matplotlib.

Q.2

A scatter plot is a type of plot used in data visualization to display the relationship between two continuous variables. It is a two-dimensional plot in which each point on the plot represents the values of the two variables.

In a scatter plot, one variable is plotted on the horizontal axis and the other variable is plotted on the vertical axis. Each data point is represented as a dot on the plot. The position of the dot on the plot represents the values of the two variables for that data point.

Scatter plots are commonly used to identify patterns or trends in data, to identify outliers or anomalies in data, and to visualize the relationship between two variables. They can also be used to visually compare two sets of data, or to explore the relationship between multiple variables by using color or size to represent additional variables.

Scatter plots are often used in fields such as finance, economics, biology, engineering, and social sciences to analyze and visualize data. Matplotlib is one of the popular libraries in Python that can be used to create scatter plots, among other types of plots.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.figure(figsize=(7,3))
plt.scatter(x,y)
plt.xlabel('This is x lable')
plt.ylabel('this is y lable')
plt.title('this is graph title')
```

Q.3

The subplot() function in Matplotlib is used to create multiple subplots within a single figure. It allows us to create a grid of plots and plot different data or different views of the same data in each subplot.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))
axs[0,0].plot(x1, y1)
axs[0,1].plot(x2, y2)
axs[1,0].plot(x3, y4)
axs[1,1].plot(x4, y4)
```

A bar plot is a type of chart that displays rectangular bars with lengths proportional to the values they represent. Bar plots are used to compare the values of different categories or groups, and to identify trends or patterns in the data.

Bar plots are commonly used to show the frequency, counts, or percentage of categorical data, such as the number of cars sold by different manufacturers, the distribution of grades in a class, or the proportion of male and female employees in a company. They can also be used to show the comparison of values for different groups over time, such as the monthly revenue of different products or services.

Bar plots are popular because they are easy to understand and interpret, and allow for quick visual comparisons between different groups or categories. They can be created using various plotting libraries in Python, such as Matplotlib, Seaborn, or Plotly.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(4,3))
plt.bar(company,profit)
```

```python
plt.figure(figsize=(4,3))
plt.barh(company,profit)
```

A box plot, also known as a box and whisker plot, is a type of chart that displays the distribution of a dataset along a single axis. It is based on five statistical summary values: the minimum value, the first quartile (25th percentile), the median (50th percentile), the third quartile (75th percentile), and the maximum value.

The box in a box plot represents the range between the first and third quartiles, while the median is shown as a vertical line inside the box. The whiskers extend from the box to the minimum and maximum values, excluding any outliers.

Box plots are used to visualize the spread and skewness of a dataset, and to identify any potential outliers or unusual observations. They are particularly useful for comparing the distribution of data between different groups or categories, as they allow for easy visual comparisons of the quartile ranges and the spread of the data.

Box plots can be created using various plotting libraries in Python, such as Matplotlib, Seaborn, or Plotly. They are commonly used in data analysis, exploratory data analysis, and statistical inference.








```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
fig, ax = plt.subplots()
ax.boxplot([box1, box2])
ax.set_title('Box plot of box1 and box2')
plt.show()
```
